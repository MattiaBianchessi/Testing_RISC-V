## Risultati sorting

Questo documento Ã¨ utilizzato per comparare i dati raccolti con l'esecuzione di alcuni algoritmi di sorting (Bubblesort, Insertionsort, Quicksort, Heapsort) eseguiti su un MacbookAir e su RISC-V

### Confronto BubbleSort
In seguito un primo confronto tra le eseguzioni di Bubblesort

```python
import pandas as pd

myPcBubbleSortWorst = pd.read_csv(
    "SortData/Foglio 1-Pc: BubbleSort( Casi peggiori).csv", sep=";", index_col=[0])

myPcBubbleSortBest = pd.read_csv(
    "SortData/Foglio 1-Pc: BubbleSort( Casi migliori).csv", sep=";", index_col=[0])

myPcBubbleSortRnd = pd.read_csv(
    "SortData/Foglio 1-Pc: BubbleSort.csv", sep=";", index_col=[0])

myPcBubble = {'Best': myPcBubbleSortBest,
              'Worst': myPcBubbleSortWorst, 'General': myPcBubbleSortRnd}

myPcBubble['General']
```

La tabella rappresenta i tempi di esecuzione dell algoritmo bubble su MacbookAir. Nella tabella si vedono i tempi minimi, massimi e la media misurati in secondi indicizzati dal numero di elementi da ordinare.

```python
RISCBubbleSortWorst = pd.read_csv(
    "SortData/Foglio 1-RISC: BubbleSort( Casi peggiori).csv", sep=";", index_col = [0])


RISCBubbleSortBest = pd.read_csv(
    "SortData/Foglio 1-RISC: BubbleSort( Casi migliori).csv", sep=";", index_col = [0])


RISCBubbleSortRnd = pd.read_csv(
    "SortData/Foglio 1-RISC: BubbleSort.csv", sep=";", index_col = [0])


RISCBubble = {'Best': RISCBubbleSortBest,
              'Worst': RISCBubbleSortWorst, 'General': RISCBubbleSortRnd}
RISCBubble['General'].head()
```

### Confronto grafico BubbleSort

In seguito i tre grafici dei tempi di computazione dell' algoritmo bubblesort eseguito sul mio pc e su RISC.

```python
import matplotlib.pyplot as plt

name =  RISCBubble['Best'].index
value00 = myPcBubble['General']['MAX']
value01 = RISCBubble['General']['MAX']

value10 = myPcBubble['General']['MIN']
value11 = RISCBubble['General']['MIN']

value20 = myPcBubble['General']['MEDIA']
value21 = RISCBubble['General']['MEDIA']

fig, ax = plt.subplots(1, 3, figsize=(16, 4), sharey=True)
ax[0].plot(name, value00, label='PC', marker = 'o')
ax[0].plot(name, value01, label = 'RISC', marker = 'o')
ax[0].set_title("Max")
ax[0].legend(loc='upper left')
ax[0].grid(True)

ax[1].plot(name, value10, label='PC', marker = 'o')
ax[1].plot(name, value11, label='RISC', marker = 'o')
ax[1].set_title("Min")
ax[1].legend(loc='upper left')
ax[1].grid(True)

ax[2].plot(name, value20, label='PC', marker = 'o')
ax[2].plot(name, value21, label='RISC', marker = 'o')
ax[2].set_title("Media")
ax[2].legend(loc='upper left')
ax[2].grid(True)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(name, RISCBubble['Best']['3'], label='Best', marker='o')
ax.plot(name, RISCBubble['Worst']['3'], label='Worst', marker='o')
ax.plot(name, RISCBubble['General']['MEDIA'], label='MEDIA', marker='o')
plt.legend(loc = 'upper left')
ax.set_title("Bubble")
plt.grid()
plt.show()
```

Il grafico rappresenta rappresenta i tempi di esecuzione di bubble sort su RISC-V. Il grafico mostra i tempi di esecuzione nei casi in cui l'array sia gia in ordine(Colore blu), in oridine inverso(colore giallo) e il caso generale(colore verde). 

```python
PCHeap = pd.read_csv(
    "SortData/Foglio 1-Pc: HeapSort.csv", sep=";", index_col=[0])
PCInsert = pd.read_csv(
    "SortData/Foglio 1-Pc: InsertionSort.csv", sep=";", index_col=[0])
PCQuick = pd.read_csv(
    "SortData/Foglio 1-Pc: QuickSort.csv", sep=";", index_col=[0])

dataPc = {"Bubble": myPcBubbleSortRnd,
            "Insertion": PCInsert, "Quick": PCQuick, "Heap":  PCHeap}

RISCHeap = pd.read_csv(
    "SortData/Foglio 1-RISC: HeapSort.csv", sep=";", index_col=[0])
RISCInsert = pd.read_csv(
    "SortData/Foglio 1-RISC: InsertionSort.csv", sep=";", index_col=[0])
RISCQuick = pd.read_csv(
    "SortData/Foglio 1-RISC: QuickSort.csv", sep=";", index_col=[0])

dataRISC = {"Bubble": RISCBubbleSortRnd,
            "Insertion": RISCInsert, "Quick": RISCQuick, "Heap":  RISCHeap}

data = {"myPC" : dataPc, "RISC": dataRISC}

#data['myPC']['Insertion'].head()
```

## Grafici a confronto


```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

ax[0].plot(name, data['myPC']['Bubble']['MEDIA'], label='PC', marker='o')
ax[0].plot(name, data['RISC']['Bubble']['MEDIA'], label='RISC', marker='o')
ax[0].set_title("Bubble sort")
ax[0].legend(loc='upper left')
ax[0].grid(True)

ax[1].plot(name, data['myPC']['Insertion']['MEDIA'], label='PC', marker='o')
ax[1].plot(name, data['RISC']['Insertion']['MEDIA'], label='RISC', marker='o')
ax[1].set_title("Insertion sort")
ax[1].legend(loc='upper left')
ax[1].grid(True)


plt.show()
```

Confronti Bubble e Insertion

```python
fig, ax = plt.subplots()
ax.plot(name, data['RISC']['Bubble']['MEDIA'], label='Bubble', marker='o')
ax.plot(name, data['RISC']['Insertion']['MEDIA'], label='Insertion', marker='o')
ax.plot(name, data['RISC']['Quick']['MEDIA'], label='Quick', marker='o')
ax.plot(name, data['RISC']['Heap']['MEDIA'], label='Heap', marker='o', color = 'red')

ax.grid(True)
plt.legend(loc='upper left')
plt.show()
```

Confronto tutti i quattro algoritmi su RISC-V. Gli algoritmi Quick e Heap sono sovrapposti, in seguito visualizzo in maniera isolata

```python
fig, ax = plt.subplots()

ax.plot(name, data['RISC']['Quick']['MEDIA'], label='Quick', marker = 'o', color = 'green')
ax.plot(name, data['RISC']['Heap']['MEDIA'], label='Heap' , marker='o', color = 'red')

ax.grid(True)
plt.legend(loc='upper left')
plt.show()
```

## Heap e Quick sort

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

ax[0].plot(name, data['myPC']['Heap']['MEDIA'], label='PC', marker='o')
ax[0].plot(name, data['RISC']['Heap']['MEDIA'], label='RISC', marker='o', color = 'red')
ax[0].set_title("Heap sort")
ax[0].legend(loc='upper left')
ax[0].grid(True)

ax[1].plot(name, data['myPC']['Quick']['MEDIA'], label='PC', marker='o')
ax[1].plot(name, data['RISC']['Quick']['MEDIA'], label='RISC', marker='o', color = 'green')
ax[1].set_title("Quick sort")
ax[1].legend(loc='upper left')
ax[1].grid(True)


plt.show()
```

## Confronto con valori raspberry pi

```python
BubbleRB = pd.read_csv('SortData/RB:BubbleSort.csv', decimal = ',', sep = ';')


InsertionRB = pd.read_csv('SortData/RB:InsertionSort.csv', decimal=',', sep=';')


HeapRB = pd.read_csv('SortData/RB:HeapSort.csv', decimal = ',', sep = ';')


QuickRB = pd.read_csv('SortData/RB:QuickSort.csv', decimal=',', sep=';')

SortingRB = {'Bubble': BubbleRB.transpose(), 'Insertion': InsertionRB.transpose(
), 'Heap': HeapRB.transpose(), 'Quick': QuickRB.transpose()}


fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)

ax[0].plot(name, SortingRB['Bubble']['MEDIA'],
           label='RB', marker='o', color='green')
ax[0].plot(name, data['RISC']['Bubble']['MEDIA'], label='RISC', marker='o')
ax[0].set_title("Bubble sort")
ax[0].legend(loc='upper left')
ax[0].grid(True)

ax[1].plot(name, SortingRB['Insertion']['MEDIA'],
           label='RB', marker='o', color='green')
ax[1].plot(name, data['RISC']['Insertion']['MEDIA'], label='RISC', marker='o')
ax[1].set_title("Insertion sort")
ax[1].legend(loc='upper left')
ax[1].grid(True)

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 4), sharey=True)
ax[0].plot(name, SortingRB['Heap']['MEDIA'],
           label='RB', marker='o', color='green')
ax[0].plot(name, data['RISC']['Heap']['MEDIA'], label='RISC', marker='o')
ax[0].set_title("Insertion sort")
ax[0].legend(loc='upper left')
ax[0].grid(True)

ax[1].plot(name, SortingRB['Quick']['MEDIA'],
           label='RB', marker='o', color='green')
ax[1].plot(name, data['RISC']['Quick']['MEDIA'], label='RISC', marker='o')
ax[1].set_title("Insertion sort")
ax[1].legend(loc='upper left')
ax[1].grid(True)

plt.show()
```
